```python
oil = pd.read_csv("/Users/alexanderkell/Documents/SGI/Projects/11-starter-kits/data/raw/starter-kits/kenya/Table5_KEN.csv")
oil
```

```python
oil = oil.pivot(index='Technology', columns='Parameter', values='Value')
oil
```

```python
oil["RegionName"] = "kenya"
oil["Time"] = "2020"
oil["Level"] = "fixed"
oil["cap_exp"] = 1
oil["fix_exp"] = 1
oil["var_par"] = 0
oil["var_exp"] = 1
oil["Type"] = "energy"
oil["EndUse"] = "oil"
oil["Agent2"] = 1
oil["InterestRate"] = 0.1
oil["MaxCapacityAddition"] = 20
oil["MaxCapacityGrowth"] = 20
oil["TotalCapacityLimit"] = 20

oil = oil.reset_index()
oil
```

```python
oil_renamed = oil.rename(columns={"Capital Cost ($/kW in 2020)": "cap_par", "Operational Life (years)": "TechnicalLife", "Technology": "ProcessName"})
oil_renamed
```

```python
oil_renamed = oil_renamed.drop(columns="Output Ratio")
oil_renamed
```
